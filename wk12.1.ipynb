# Z score method

```python
sns.boxplot(data['math score'])
```

```python
sns.distplot(data['math score'])
```

```python
upper_limit=data['math score'].mean()+3*data['math score'].std()
lower_limit=data['math score'].mean()-3*data['math score'].std()
len(data['math score'])
```

```python
new_data=data.loc[(data['math score']<upper_limit)&(data['math score']>lower_limit)]
new_data.head()
len(new_data)
```

```python
sns.boxplot(new_data['math score'])
```

```python
sns.distplot(new_data['math score'])
```

# IQR method

```python
data.head()
```

```python
q1=data['math score'].quantile(.25)
q3=data['math score'].quantile(.75)
IQR=q3-q1
q1,q3,IQR
```

```python
upper_limit=q3+(1.5*IQR)
lower_limit=q1-(1.5*IQR)
upper_limit,lower_limit
```

```python
sns.distplot(data['math score'])
```

```python
sns.boxplot(data['math score'])
```

```python
new_data=data.loc[(data['math score']<upper_limit)&(data['math score']>lower_limit)]
```

```python
sns.boxplot(new_data['math score'])
```

```python
sns.distplot(new_data['math score'])
```

```python
q1=data['reading score'].quantile(.25)
q3=data['reading score'].quantile(.75)
IQR=q3-q1
q1,q3,IQR
```

```python
upper_limit=q3+(1.5*IQR)
lower_limit=q1-(1.5*IQR)
upper_limit,lower_limit
```

```python
new_data=data.loc[(data['reading score']<upper_limit)&(data['reading score']>lower_limit)]
len(new_data)
```

```python
sns.boxplot(new_data['reading score'])
```

```python
sns.distplot(new_data['reading score'])
```

# Correlation


```python
new_data.head()
```

```python
correlation=new_data[['gender','math score']].corr()
# sns.heatmap(correlation,xticklabel='math score',yticklabel='gender',annot=True)
```

```python
new_data['gender_encoded']=new_data['gender'].apply(lambda x: 1 if x=='male' else 0)
```

```python
correlation_maths=new_data['gender_encoded'].corr(new_data['math score'])
# correlation_maths
if correlation_maths>0:
    print('males may have slightly higher math scores ')
elif correlation_maths<0:
    print('females may have slightly higher math scores')
```

```python
correlation_writting=new_data['gender_encoded'].corr(new_data['writing score'])
# correlation_writting
if correlation_writting>0:
    print('males may have slightly higher writing scores ')
elif correlation_writting<0:
    print('females may have slightly higher writing scores')
```

```python
correlation_reading=new_data['gender_encoded'].corr(new_data['reading score'])
# correlation_reading
if correlation_reading>0:
    print('males may have slightly higher reading scores ')
elif correlation_reading<0:
    print('females may have slightly higher reading scores')
```

```python
correlation=new_data[['math score','reading score','writing score']].corr()
correlation
```

```python
sns.heatmap(correlation,annot=True)
```
